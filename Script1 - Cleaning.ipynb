```python
import pandas as pd
import numpy as np
import os
```

```python
pwd = os.getcwd()
```

```python
toyota = pd.read_excel(pwd + "\\Car_Selection_2021.xlsx", sheet_name = "Toyota")
honda = pd.read_excel(pwd + "\\Car_Selection_2021.xlsx", sheet_name = "Honda")
subaru = pd.read_excel(pwd + "\\Car_Selection_2021.xlsx", sheet_name = "Subaru")
mazda = pd.read_excel(pwd + "\\Car_Selection_2021.xlsx", sheet_name = "Mazda")
nissan = pd.read_excel(pwd + "\\Car_Selection_2021.xlsx", sheet_name = "Nissan")
ford = pd.read_excel(pwd + "\\Car_Selection_2021.xlsx", sheet_name = "Ford")
chevrolet = pd.read_excel(pwd + "\\Car_Selection_2021.xlsx", sheet_name = "Chevrolet")
kia = pd.read_excel(pwd + "\\Car_Selection_2021.xlsx", sheet_name = "Kia")
lexus = pd.read_excel(pwd + "\\Car_Selection_2021.xlsx", sheet_name = "Lexus")
mercedes = pd.read_excel(pwd + "\\Car_Selection_2021.xlsx", sheet_name = "Mercedes-Benz")
```

```python
toyota_mod = toyota.copy()
honda_mod = honda.copy()
subaru_mod = subaru.copy()
mazda_mod = mazda.copy()
nissan_mod = nissan.copy()
ford_mod = ford.copy()
chevrolet_mod = chevrolet.copy()
kia_mod = kia.copy()
lexus_mod = lexus.copy()
mercedes_mod = mercedes.copy()
```

```python
columns_to_drop = ['Net Torque', 'Turning Diameter (ft)', 'Transmission', 'MPGe (combined)', 'MPGe (city)', 'MPGe (highway)', 'Electricity Range (mi)', 'KWh/100 Mi']
```

```python
toyota_mod.drop(columns_to_drop, inplace = True, axis = 1)
honda_mod.drop(columns_to_drop, inplace = True, axis = 1)
subaru_mod.drop(columns_to_drop, inplace = True, axis = 1)
mazda_mod.drop(columns_to_drop, inplace = True, axis = 1)
nissan_mod.drop(columns_to_drop, inplace = True, axis = 1)
ford_mod.drop(columns_to_drop, inplace = True, axis = 1)
chevrolet_mod.drop(columns_to_drop, inplace = True, axis = 1)
kia_mod.drop(columns_to_drop, inplace = True, axis = 1)
lexus_mod.drop(columns_to_drop, inplace = True, axis = 1)
mercedes_mod.drop(columns_to_drop, inplace = True, axis = 1)
```

```python
chevrolet_mod.head()
```

```python
# fun cell to count number of cars in a specific classification for Toyota
count = toyota_mod[toyota_mod["EPA Classification"].notna()]
count = count.groupby(["EPA Classification"])["Name"].nunique().reset_index()
count.rename(columns = {"Name" : "Number of Cars"}, inplace = True)
count
```

```python
combined = pd.concat([toyota_mod, honda_mod, subaru_mod, mazda_mod, nissan_mod, ford_mod, chevrolet_mod, kia_mod, lexus_mod, mercedes_mod])
```

```python
combined_mod = combined.copy()
```

```python
# checks for empty strings and replaces with NaN
combined_mod.replace(r'^\s*$', np.nan, regex = True, inplace = True)
```

```python
combined_mod
```

```python
class_count = combined_mod.groupby(["EPA Classification"])["Name"].nunique().reset_index()
class_count.rename(columns = {"Name" : "Number of Cars"}, inplace = True)
class_count
```

```python
drivetrain_count = combined_mod.groupby(["Drivetrain"])["Name"].nunique().reset_index()
drivetrain_count.rename(columns = {"Name" : "Number of Cars"}, inplace = True)
drivetrain_count
```

```python
engine_count = combined_mod.groupby(["Engine Type"])["Name"].nunique().reset_index()
engine_count.rename(columns = {"Name" : "Number of Cars"}, inplace = True)
engine_count
```

```python
cylinder_count = combined_mod.groupby(["Cylinder Layout"])["Name"].nunique().reset_index()
cylinder_count.rename(columns = {"Name" : "Number of Cars"}, inplace = True)
cylinder_count
```
